```python
#What is the average age for male versus female Titanic passengers?
titanic[["Sex", "Age"]].groupby("Sex").mean()
# also can be written as
titanic.groupby("Sex")["Age"].mean()
```

```python
# - To get the average age by gender:  
#   1. Select columns: `titanic[["Sex", "Age"]]`.  
#   2. Group by `Sex`: use `groupby("Sex").mean()`.  
# - Grouping follows the split-apply-combine pattern:  
#   1. **Split**: Divide data into groups.  
#   2. **Apply**: Apply a function to each group.  
#   3. **Combine**: Merge results into a structure.
```

```python
# What is the number of passengers in each of the cabin classes?
titanic["Pclass"].value_counts()
# also can be written as
titanic.groupby("Pclass")["Pclass"].count()
```

```python
# - **`size` vs `count`**:  
#   - `size`: Includes `NaN`, counts all rows.  
#   - `count`: Excludes `NaN`.  
# - **`value_counts`**: Use `dropna` to include/exclude `NaN`.  
# - **Key Points**:  
#   - Aggregate stats work on full columns/rows.  
#   - `groupby`: Uses the split-apply-combine pattern.  
#   - `value_counts`: Shortcut to count entries per category.
```
