```python
                                 ###-- IMPORT PANDAS --###
# **Pandas** is a powerful open-source Python library for data analysis and manipulation. 
# It provides two main data structures: **DataFrame** (2D tabular data) and **Series** (1D array). 
# Pandas makes it easy to clean, transform, and analyze data using intuitive functions.
# It is widely used in data science and machine learning workflows for tasks like data wrangling and preprocessing.
```

```python
import pandas as pd #Syntax to Import pandas in you file. [Install pandas using "https://pandas.pydata.org/docs/getting_started/index.html"]
```

```python
# Pandas stores data in table form which are called dataFrames. (DataFrames can be compared to spreadSheets).
df = pd.DataFrame(
        {
            "Name": [
                "Braund, Mr. Owen Harris",
                "Allen, Mr. William Henry",
                "Bonnell, Miss. Elizabeth",
            ],
            "Age": [22, 35, 58],
            "Sex": ["male", "male", "female"],
        }
    )
df
```

```python
# DataFrame is a 2-D Data Structure. It can store 
# characters, integers, floating point values, categorical data and more
```

```python
# Each column is a DataFrame is a SERIES.
# Series :- A pandas Series has no column labels, as it is just a single column of a DataFrame. A Series does have row labels.

df["Age"] #Age col is a series and not Df now.
```

```python
# Can create a series from scratch as well.
ages = pd.Series([22, 35, 58], name="Age")
ages
```

```python
# max() function on ages column
df["Age"].max()
# or
ages.max()
```

```python
# describe() function gives some basic stats.
df.describe()
```
